```python
# author: github.com/petr0leum/Numerical_analysis

from numpy import pi, sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = pi/7
arr = np.array([0.5 * h, 1.5 * h, 2.5 * h,4.5 * h, 6.5 * h])
linspace = np.linspace(0, 2*pi, 150)
```

```python
F = lambda x: np.sin(x) * sqrt(x) + 1
Phi = lambda x, i: np.prod(np.array([((x-arr[j])/(arr[i]-arr[j]))
                                for j in range(0, len(arr)) if j!=i]), axis=0)
BasePoly = lambda x, i: np.array([F(arr[i])*Phi(x,i)]).transpose()
Lagrange = lambda: np.sum([BasePoly(linspace, i) for i in range(0, len(arr))], axis=0)
```

```python
# Design:
ax = plt.subplots(figsize=(16, 8))
colors = ["#00FA9A", "#00FFFF", "#FF4500", "#008000", "#00008B"]
ax = plt.gca()
plt.xlabel(r'$linspace\ \ X\ \in\ {\rm I\!R}$')
plt.xlim(0, 2*pi); plt.ylim(-2, 4); plt.title("Lagrange polynomial")
# Plotting
plt.plot(linspace, F(linspace), linewidth=3, color="#525c5d", label=r'$F(x)$')
for i in range(0, len(arr)):
    plt.scatter(arr[i], F(arr[i]), s=150, color="#1C4C89", marker="+")
    plt.plot(linspace, BasePoly(linspace, i), color=colors[i], label=r'$Base\ L_{}$'.format(i+1))
plt.scatter(linspace, Lagrange(), s=3.5, color="#1C4C89")
ax.legend()
plt.show()
```
