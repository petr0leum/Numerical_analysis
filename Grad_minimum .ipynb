```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
function = lambda x, y: 1 - np.cos(x) * np.cos(y)
x_dots = np.linspace(-5, 5, 100)
y_dots = np.linspace(-5, 5, 100)
X_space, Y_space = np.meshgrid(x_dots, y_dots, sparse = True)
Z = function(X_space, Y_space)
```

```python
def iterative_method(x, y, k, h = 0.1):
    xlist = [x]
    ylist = [y]
    xk = x
    yk = y
    for i in range(k):
        xk = xk - h * np.sin(xk) * np.cos(yk)
        yk = yk - h * np.cos(xk) * np.sin(yk)
        xlist.append(xk)
        ylist.append(yk)
    print(xk, yk)
    fig = plt.figure()
    ax3D = fig.add_subplot(111, projection='3d')
    ax3D.plot_surface(X_space, Y_space, Z, alpha=0.8)  
    ax3D.scatter3D(xlist, ylist, function(xlist, ylist), s = 40, c = 'r')
    ax3D.set_xlabel("x")
    ax3D.set_ylabel("y")
    ax3D.set_zlabel("z")
iterative_method(0, 3, 640)
```

```python
def difference_method(x, y, k, f = function, h = 0.1):
    xlist = [x]
    ylist = [y]
    for i in range(k):
        x -= f(x + h/2, y) - f(x - h/2, y)
        xlist.append(x)
        y -= f(x, y + h/2) - f(x, y - h/2)
        ylist.append(y)

    fig = plt.figure()
    ax3D = fig.add_subplot(111, projection='3d')
    ax3D.plot_surface(X_space, Y_space, Z, alpha=0.6)  
    ax3D.scatter3D(xlist, ylist, f(xlist, ylist), s = 40, c = 'r')
    ax3D.set_xlabel("x")
    ax3D.set_ylabel("y")
    ax3D.set_zlabel("z")
difference_method(0, 3, 640)
```
