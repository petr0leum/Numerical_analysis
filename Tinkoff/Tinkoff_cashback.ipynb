```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Tinkoff = pd.read_csv('C:\\Users\\pet-1\\p\\Homework\\Tinkoff\\Tinkoff_kopia.csv', encoding='windows - 1251', sep=';' )
```

```python
Tinkoff
```

```python
Tinkoff.columns = ['Date1', 'Date2', 'Card number','Status','Sum oper','Cur oper','Sum pay','Cur pay','Cashback','Category'
                   ,'MCC','Description','Bonus']
```

```python
Tinkoff['Cashback'].plot(color = 'g', figsize=(10, 6))
plt.title('Cashback')
plt.grid()
plt.show()
```

```python
Tinkofff = Tinkoff['Category'].value_counts()
Tinkofff[:15]
Tinkofff[:15].plot(kind='bar', title='Частые категории', figsize=(10, 6))
plt.show()
```

```python
Tinkofff = Tinkoff['Description'].value_counts()
Tinkofff[:15]
Tinkofff[:15].plot(kind='pie', title='Топ описаний', figsize=(10, 6))
plt.show()
```

```python
Tinkoff.groupby(['MCC','Category'])['Cashback'].count()
Tinkoff.groupby(['MCC','Category'])['Cashback'].aggregate(sum)
```
